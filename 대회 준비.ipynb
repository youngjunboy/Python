```python
import pandas as pd
import matplotlib.pyplot as plt

k = pd.read_csv('민석이와 친구들-데이터.csv', encoding='cp949')
k.head()
```

```python
k['time'].value_counts().plot(kind='bar')
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.title('오전 VS 오후')
plt.show()
```

```python
k[['sex','time']].value_counts().plot(kind='bar')
```

```python
k['week_kyo'].value_counts().plot(kind='bar')
plt.xticks(range(5), ["MON","TUE","WED","THU","FRI"])
plt.show
```

```python
k['week_kyo'].value_counts().plot(kind='pie')
plt.show
```
