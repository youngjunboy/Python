```python
import csv
import matplotlib.pyplot as plt

f = open("subway202306.csv")
data = csv.reader(f)
next(data)
next(data)

mx = 0
mx_station = ''

t1 = input("지하철 승하차 시작시간 입력 : ")
t2 = input("지하철 승하차 마지막시간 입력 : ")

a = 2*int(t1)-4    #2x-4
b = 2*int(t2)-3    #2x-3 왜냐하면 인덱스 슬라이싱 할 때 그전까지 나오므로 플러스 1을 해주면 -3이 된다.

print(a)
print(b)

for row in data:
    row[4:] = map(int, row[4:])
    if sum(row[a:b:2]) > 300000:
        mx = sum(row[a:b:2])
        mx_station = row[3] + "(" + row[1] + ")"
        print(mx_station, mx)
```

```python
import csv
import matplotlib.pyplot as plt

f = open("subway202306.csv")
data = csv.reader(f)
next(data)
next(data)

mx = ['0'] * 23
mx_station = [''] * 23

for row in data:
    row[4:] = map(int, row[4:])
    for i in range(23):
        a = row[2*i+4]
        if a > int(mx[i]):
            mx[i] = a
            mx_station[i] = row[3]
                
plt.rc("font",family = "Malgun Gothic")
plt.bar(range(23), mx)
plt.xticks(range(23), mx_station, rotation=90)
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt
import numpy as np

f = open("pop202306.csv")
data = csv.reader(f)
next(data)
data = list(data)

name1 = input("첫번째 지역이름 입력 : ")
name2 = input("두번째 지역이름 입력 : ")

for row in data:
    if name1 in row[0]:
        for i in row[3:]:
            home1 = np.array(row[3:], dtype=int)
        break
        
for row in data:
    if name2 in row[0]:
        for i in row[3:]:
            home2 = np.array(row[3:], dtype=int)
        break
    
plt.rc("font", family="Malgun Gothic")
plt.rcParams["axes.unicode_minus"]=-False
plt.title(name1+ name2)
plt.plot(home1)
plt.plot(home2)
plt.show()
```

# *** 아래 소스코드 중요

```python
import csv
import matplotlib.pyplot as plt
import numpy as np

f = open("pop202306.csv")
data = csv.reader(f)
next(data)
data = list(data)

name = input("지역 이름을 입력 : ")
mn = 1
result_name = ""
result = 0
plt.rc("font", family="Malgun Gothic")

for row in data:
    if name in row[0]:
        if int(row[2]) != 0:
            for i in row[3:]:
                home = np.array(row[3:], dtype=int)/int(row[2])
        
for row in data:
    if int(row[2]) != 0:
        away = np.array(row[3:], dtype=int)/int(row[2])
        s = np.sum((home-away)**2)
        if s < mn and name not in row[0]:
            mn = s
            result_name = row[0]
            result = away
            
plt.style.use("ggplot")
plt.figure(figsize=(10,5), dpi=300)
plt.title(name + "지역과 가장 비슷한 인구구조를 가진 지역")
plt.plot(home, label=name)
plt.plot(result, label=result_name)
plt.legend()
plt.show()
```
