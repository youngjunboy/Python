```python
import pandas as pd
df = pd.read_csv("data/employees.csv")
df.head(10)
```

```python
emp=pd.read_csv("employees.csv", parse_dates = ["Start Date", "Last Login Time"])
emp["Senior Management"] = df["Senior Management"].astype("bool")
emp["Gender"] = df["Gender"].astype("category")
```

```python
mask1 = emp["Gender"] == "Male"
mask2 = emp["Team"] == "Marketing"

emp[mask1 & mask2]
```

```python
mask1 = emp["Gender"] == "Female"
mask2 = emp["Team"] == "Finance"
emp[mask1 & mask2]
```

```python
mask1 = emp["Gender"] == "Male"
mask2 = emp["Team"] == "Client Services"
mask3 = emp["Start Date"] > "2016-06-01"
emp[(mask1 & mask2) | mask3]
```

```python
mask1 = emp["Team"] == "Legal"
mask2 = emp["Team"] == "Sales"
mask3 = emp["Team"] > "Product"
emp[mask1 | mask2 | mask3]
```

```python
mask = emp["Team"].isnull()
emp[mask]
#isnull = isna
```

```python
emp[emp["Gender"].notnull()]
```

```python
mask1 = emp["Salary"]>=60000
mask2 = emp["Salary"]<=70000
emp[mask1 & mask2]
```

```python
emp[emp["Salary"].between(60000, 70000)]
```

```python
emp[emp["Bonus %"].between(2,5)]
```
