```python
import csv

f = open("data/age_final.csv")
data = csv.reader(f)
header = next(data)
result = []
for row in data:
    if "신도림" in row[0]:
        for i in row[3:13]:
            result.append(int(i))
sum(result)
```

```python
import csv
import matplotlib.pyplot as plt
f = open("data/age_final.csv")
data = csv.reader(f)
header = next(data)
b = []
for row in data:
    if "후암동" in row[0]:
        for i in row[3:]:
            b.append(int(i))
plt.plot(b, label="연령별분포",color = "blue")
plt.rc("font", family = "Malgun Gothic")
plt.title("후암동 인구분포")
plt.legend()
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt
f = open("data/age_final.csv")
data = csv.reader(f)
header = next(data)
a = input("지역을 입력하세요 : ")
b = []
for row in data:
    if a in row[0]:
        for i in row[3:]:
            b.append(int(i))
plt.plot(b, label="연령별분포", color = "hotpink")
plt.rc("font", family = "Malgun Gothic")
plt.title(a+" 인구분포")
plt.legend()
plt.show()
```
