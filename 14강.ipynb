```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
salaries = pd.read_csv("data/Salaries.csv", low_memory=False)
```

```python
plt.style.use('fivethirtyeight')

salaries["BasePay"] = pd.to_numeric(salaries["BasePay"], errors="coerce").fillna(0)
salaries["OvertimePay"] = pd.to_numeric(salaries["OvertimePay"], errors="coerce").fillna(0)
salaries["OtherPay"] = pd.to_numeric(salaries["OtherPay"], errors="coerce").fillna(0)
df = salaries[["EmployeeName", "BasePay", "OvertimePay", "OtherPay"]]
df.set_index("EmployeeName", inplace=True)
```

```python
df.hist(bins=100)
```

```python
df.plot(kind="box")
```

```python
ufos.date_time.dt.month.value_counts().sort_index().plot(kind = "bar")
```

```python
month = {1:"Jan",2:"Feb",3:"Mar",4:"Apr",5:"May",6:"Jun",7:"Jul",8:"Aug",9:"Sep",10:"Oct",11:"Nov",12:"Dec",}
# 딕셔너리로 함수를 만들어서 아래 rename 함수에다가 집어넣은다. 그럼 x축 이름들이 바뀜
```

```python
fig, axs = plt.subplots(2,3,figsize=(14,8))
fig.suptitle("UFO sightings by month", fontsize=14)
fig.tight_layout(pad=2)
# plt.tight_layout()
ufos[ufos.date_time.dt.year==2014].date_time.dt.month.value_counts().sort_index().rename(month).plot(kind="bar",ax=axs[0][0],title="2014")
ufos[ufos.date_time.dt.year==2015].date_time.dt.month.value_counts().sort_index().rename(month).plot(kind="bar",ax=axs[0][1],title="2015")
ufos[ufos.date_time.dt.year==2016].date_time.dt.month.value_counts().sort_index().rename(month).plot(kind="bar",ax=axs[0][2],title="2016")
ufos[ufos.date_time.dt.year==2017].date_time.dt.month.value_counts().sort_index().rename(month).plot(kind="bar",ax=axs[1][0],title="2017")
ufos[ufos.date_time.dt.year==2018].date_time.dt.month.value_counts().sort_index().rename(month).plot(kind="bar",ax=axs[1][1],title="2018")
ufos[ufos.date_time.dt.year==2019].date_time.dt.month.value_counts().sort_index().rename(month).plot(kind="bar",ax=axs[1][2],title="2019")
# ufos의 year이 2014 ~ 2019와 같은 것의 value_coounts()개수를 측정해서 sort_index로 정렬하고 rename()으로 이름 바꾸기.
```
